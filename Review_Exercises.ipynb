# 2)	Write a Python Program and perform the following operations

I.    Create a List {1225, 4986, 6789, 7890, 2345, 6783, 0987, 1234, 8765, 3456}

````python
numberList = [1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]\

print(numberList)
````

II.    Iterate using a for loop

````python
for number in numberList:
    print(number)
````

III.    Iterate using for loop and range

````python
for i in range(len(numberList)):
    print(numberList[i])
````

IV.    List Comprehension

````python
new_list = [number for number in numberList]

print(new_list)
````

V.    Enumerate

````python
for index, number in enumerate(numberList):
    print(index, number)
````

VI.    Iter function and next function

````python
iterator = iter(numberList)

while True:
    try:
        number = next(iterator)
        print(number)
    except StopIteration:
        break
````

VII.    Map function

````python
def square(number):
    return number ** 2

squared_numbers = map(square, numberList)

print(list(squared_numbers))
````

VIII.    Using zip

````python
numberList1 = [1, 2, 3, 4, 5]
numberList2 = [10, 20, 30, 40, 50]

for num1, num2 in zip(numberList1, numberList2):
    print(num1, num2)
````

IX.    Using NumPy Module

````python
import numpy as np

numberArray = np.array(numberList)

print(numberArray)
````

# 3)	For a List A, B, C, D, E write a python program to compute all the combinations and permutations 

````python
def combinations(lst):
  if not lst:
    return [[]]
  return [[lst[0]] + comb for comb in combinations(lst[1:])] + combinations(lst[1:])

def permutations(lst):
  if len(lst) <= 1:
    return [lst]
  perms = []
  for i in range(len(lst)):
    perms.extend([[lst[i]] + p for p in permutations(lst[:i] + lst[i+1:])])
  return perms

list = ["A", "B", "C", "D", "E"]

print("Combinations\n")
for combination in combinations(list):
    print(combination)

print("\n\n")
print("Permutations\n")
for permutation in permutations(list):
    print(permutation)
````

# 4)	Using the same list use itertools compute permutations and combinations 

````python
import itertools

def combinations(elements):
    combinations = []

    for length in range(1, len(elements) + 1):
        combinations.extend(itertools.combinations(elements, length))

    return combinations

def permutations(elements):
    permutations = []

    for length in range(1, len(elements) + 1):
        permutations.extend(itertools.permutations(elements, length))

    return permutations


list = ["A", "B", "C", "D", "E"]

print("Combinations\n")
for combination in combinations(list):
    print(combination)

print("\n\n")
print("Permutations\n")
for permutation in permutations(list):
    print(permutation)
````
